```python
#export
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def pdfs(time_box:int=30, 
         purpose:str='Inputs_Outputs_WhoWhatWhyWhenWhereHow, making it TRUE ADEPTive', 
         design:str='STEP1. STEP2. STEP3',
         feedback:str='What goes well? What did not? What to change?',
         strength:int=5,
         csv_file:str='drive/My Drive/vblog/_notebooks/data/'+'2022-08-21_08-27.csv',
         last_nrows:int=3,
         ):
  """Q: How do you consistently focus the principle goals (eg. GYMi) in these 3 months (07/2022 - 10/2022)? 
  A: do PDFS on each & every time-box
  
  Args:
  - Purpose: what you want to achieve within this time-box (e.g. 3 bullet points each can be done within ~10 minutes)

  - Desgin: Who are doing What for Why by How under When (logical condition) & Where (story context), visualized as a dynamical graph neural network

  - Feedback: strategic feedback data collection, analysis and recommendation for people, process, product upgrade

  - Strength: measurement of both qualitative and quantitative metrics for objective function

  - csv_file:str='drive/My Drive/vblog/_notebooks/data/'+'2022-08-21_08-27.csv',
  Returns:
  - an interactive dashboard visualizing both the gains and the gaps
  - recommendation for the next best steps in order to achieve the optimal weekly objectives
  Examples:
    pdfs(time_box=30, 
    purpose='1.why duplicated googel_event_id 2. how to use model trained by 100k to make prediction of 1M data', 
    design='STEP1.debug STEP2.reload 1M data & prep_data(1M data) STEP3.make prediction on 1M data with 100k model',
    feedback='What goes well? ez execution What did not? to good to feellike real(: What to change? rinse & repeat!!',
    strength=5,
    csv_file='drive/My Drive/vblog/_notebooks/data/'+'2022-08-21_08-27.csv',
    )

    #seer model
    story ='1.collect & save data; 2. clean up & visualize data; 3. EDA; 4.smart recommendations' 
    expectation='Paul Wu is creating this pdfs function for the sake of 12X growth by DS study on each 30 min time boxing at home'
    execution='define inputs & outputs & then work out every layer between'
    reward=5
  """


  from datetime import datetime, timedelta
  from pytz import timezone

  # define date format
  dt_format = '%Y-%m-%d %H:%M'

  # datetime object containing current date and time
  start_dt = datetime.now(tz=timezone('America/New_York'))
  end_dt = start_dt + timedelta(minutes=time_box)

  start, end = start_dt.strftime(dt_format), end_dt.strftime(dt_format)
  if csv_file is None: #str='2022-08-21'
    csv_file = 'drive/My Drive/vblog/_notebooks/data/' + start[:10]+'_'+end[5:10]+'.csv'

    
  # The data assigned to the dictionary 
  pdfs_dic = {'start_dt':start,
              'end_dt':end,
              'time_box':time_box,
              'purpose':purpose,
              'design':design,
              'feedback':feedback,
              'strength':strength,
              } 
  fieldnames = list(pdfs_dic.keys())  

  # To add a dictionary as a new row to an existing csv file
  # Ref: https://adamtheautomator.com/read-csv-in-python/#:~:text=If%20you%20need%20to%20append,a%20new%20CSV%20file%20(demo_csv1.
  import csv

  # In case csv file does not exist, creat it
  from pathlib import Path

  my_file = Path(csv_file)
  if not my_file.is_file(): # in case the csv file no exist, creat such a file
    with open(csv_file, 'w') as f:
      writer = csv.writer(f) #this is the writer object
      writer.writerow(fieldnames) # this will list out the names of the columns which are always the first entrries
  
  #double check if lastest task's end datetime < current start daytime:
  df = pd.read_csv(csv_file, index_col=0)
  from IPython.display import display
  print(f"=============last {last_nrows} rows before update===================")
  display(df.tail(last_nrows))
  
  from dateutil.parser import parse
  if df.shape[0]>=1: # in case df has 1+ row
    #datetime.now().timestamp() > parse(compare_date).timestamp() #REF: https://stackoverflow.com/questions/60003764/typeerror-cant-compare-offset-naive-and-offset-aware-datetimes
    if parse(df.iloc[-1,:]['end_dt']).timestamp() > parse(start).timestamp(): # in case the last time-boxing is still going on
        #print("ATTENTION: Last time_box is still running \n=========> CLOSE the old loop BEFORE starting a new one!")
        update = input('ATTENTION: The last task is still running \n=========> Do you want to update it [Y/N]?\n') 
        if ('n' in update) or ('N' in update):
          return
        else: #renew the last record & write df back to csv_file
          df.loc[df.index[-1],'purpose'] = purpose # ref:https://stackoverflow.com/questions/25160385/change-an-element-in-the-last-row-of-a-dataframe 1)df.loc[df.index[-1],"purpose"] 2)df.columns.get_loc('my_col_name')
          df.loc[df.index[-1],'design'] = design
          df.loc[df.index[-1],'feedback'] = feedback
          df.loc[df.index[-1],'strength'] = strength
          print(f"=============last {last_nrows} samples AFTER update===================\n")
          display(df.tail(last_nrows))
          df.to_csv(csv_file)        
    else: # in case the last time-boxing is done
      with open(csv_file, 'a') as f:
        dict_object = csv.DictWriter(f, fieldnames=fieldnames) 
        dict_object.writerow(pdfs_dic)
      print(f"=============lastest sample AFTER update===================")
      df = pd.read_csv(csv_file).tail(1)
      display(df)
  elif df.shape[0]==0: # in case df is empty (e.g. just has the headers but no data row)
    with open(csv_file, 'a') as f:
      dict_object = csv.DictWriter(f, fieldnames=fieldnames) 
      dict_object.writerow(pdfs_dic)
    print(f"=============lastest sample AFTER update===================")
    df = pd.read_csv(csv_file).tail(1)
    display(df)
  else:
    return

```

#### demo: 10X by exponential growth

```python
#export

from ipywidgets import interact, interactive, IntSlider, Layout, FloatSlider
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive, FloatSlider, IntSlider
from IPython.display import display, Image



def plot_nx(r=0.05, n=52, N=10, iter_unit='day'):
    """visualize how you can NX your productivity
    Args:
        r (float) default= 0.05, weekly growth rate
        n (int) default= 52,  number of weeks
        N (int) default=10, the goal of N x your productivity
    """
    x = [i for i in range(n+1)] # timeline by week
    y = [(1+r)**i for i in x] # productivity outcome

    fig, ax = plt.subplots(figsize=(8,6))
    ax.scatter(x, y)
    ax.axhline(y=N)
    ax.set_xlabel(f'Number of {iter_unit}')
    ax.set_ylabel('Your Output')
    ax.set_title(f'Your accumulative output is {int(np.ceil(y[-1]))}X in {n} {iter_unit} with growth rate {r*100}%');
    
    
plot_nx(r=0.05, n=52, N=10, iter_unit='week')
```

```python
#export
plot=interactive(plot_nx,
                r=FloatSlider(
                            value=0.1,
                            min=0.01,
                            max=.9,
                            step=0.01,
                            description='growth rate:',
                            disabled=False,
                            continuous_update=False,
                            orientation='horizontal',
                            readout=True,
                            readout_format='.2f'),
                 n=IntSlider(
                            value=38,
                            min=1,
                            max=int(60),
                            step=1,
                            description='Iterations:',
                            disabled=False,
                            continuous_update=False,
                            orientation='horizontal',
                            readout=True,
                            readout_format='d'),
                N=IntSlider(
                            value=10,
                            min=1,
                            max=int(60),
                            step=1,
                            description='N-fold:',
                            disabled=False,
                            continuous_update=False,
                            orientation='horizontal',
                            readout=True,
                            readout_format='d'))
    
display(plot)
```
